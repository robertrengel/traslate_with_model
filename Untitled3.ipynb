```python
from google.colab import files
text = files.upload()
print(text)
filename = list(text.keys())[0]
print(filename)
```

```python
import PyPDF2
```

```python
pdf_file_obj = open(filename,"rb")
```

```python
pdf_reader = PyPDF2.PdfReader(pdf_file_obj)
```

```python
print(len(pdf_reader.pages))
```

```python
for page in range(3):
  pdf_pages = pdf_reader.pages[page]
  text = pdf_pages.extract_text()
  print(text)
```
