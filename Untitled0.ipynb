```python
import torch, random, textwrap
from transformers import pipeline
en_fr_translator = pipeline("translation", model="Helsinki-NLP/opus-mt-tc-big-en-es")
```

```python
from google.colab import files
text = files.upload()
print(text)
filename = list(text.keys())[0]
print(filename)
```

```python
text = text[filename].decode("utf-8")
parrafo = text.split("\n")
line_list = [line for line in parrafo if line]
filename = f"traslate_{random.randint(1, 10000)}.txt"
```

```python
with open(filename, "a") as file:
  for i in en_fr_translator(line_list):
    file.write(i['translation_text'] + "\n")
```

```python
files.download(filename)
```
