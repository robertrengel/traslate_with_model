```python
import csv
```

```python
from google.colab import files
text = files.upload()
print(text)
filename = list(text.keys())[0]
print(filename)
```

```python
# Abre el archivo .txt con los bloques de tres líneas de texto
with open(filename, "r") as archivo_txt:
    lineas = archivo_txt.readlines()
    
```

```python
# Imprime las primeras 10 líneas de la lista
for i in range(10):
    print(lineas[i])
```

```python
# Crea un archivo .csv y un objeto csv.writer
with open("archivo_data_with_header6.csv", "w", newline="") as archivo_csv:
    escritor_csv = csv.writer(archivo_csv)
    escritor_csv.writerow(["English", "Spanish"])
    # Recorre cada bloque de tres líneas en el archivo .txt

    for i in range(12000, 23998, 4):
        primera_linea = lineas[i].strip()
        tercera_linea = lineas[i + 2].strip()

        # Escribe las líneas en el archivo .csv
        escritor_csv.writerow([primera_linea, tercera_linea])
```
